```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("C:/Users/Student/Desktop/7Th Sem/dip/Pepper.tiff",0)

fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(2,2,1)
ax1.set_title('Image')
ax1.imshow(img,cmap='gray')
```

```python
plt.hist(img.ravel(),256,[0,256]);
plt.show()
```

```python
m,n = img.shape
print(m)
print(n)
```

```python
h = []
size = m*n
p = float(1.0)/size
print(p)
for i in range(256):
    h.append(0)

print(h)
```

```python
nk =[]
graylevel =[]
for i in range(m):
    for j in range(n):
        r =img[i][j]
        if r not in graylevel:
            graylevel.append(r)
graylevel.sort()

L = len(graylevel)
print(L)

for i  in range(len(graylevel)):
    nk.append(0)

for i in range(m):
    for j in range(n):
        r = img[i][j]
        pos = graylevel.index(r)
        nk[pos] = 1+ nk[pos]
        
for i in range(len(nk)):
    value = graylevel[i]
    count = nk[i]
    print(value,count)
```

```python
pdf =[]
size =m*n
print(size)
for i in range(L):
    value = (float(nk[i])/size)
    pdf.append(value)
    
print(pdf)
```

```python
sum=0
cdf = []
for i in range(len(pdf)):
    sum =sum+ pdf[i]
    cdf.append(sum)
print(cdf)
```

```python
newcdf = []
for i in cdf:
    newcdf.append(int(i*(L-1)))
for i in range(len(pdf)):
    value = graylevel[i]
    count = newcdf[i]
    print(value,count)
```

```python
newImg = np.zeros(img.shape)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        try:
            newImg[i][j] = newcdf[img[i][j]]
        except:
            continue
```

```python
plt.hist(newImg.ravel(),256,[0,256])
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2,2,1)
ax1.set_title('Image')
ax1.imshow(img,cmap='gray')
ax2 = fig.add_subplot(2,2,2)
ax2.set_title('Equalized Image')
ax2.imshow(newImg,cmap='gray')
```
